```python
class UnoAgentColorMatching(object):
    def __init__(self, num_actions):
        ''' 
        Args:
            num_actions (int): The size of the ouput action space
        '''
        self.use_raw = True
        self.num_actions = num_actions
        
    def step(self, state):
        legal_actions = state['raw_legal_actions']
        state = state['raw_obs']
        current_color = state['target'][0].split('-')[0]
        
        color_matching_actions = [action for action in legal_actions if action.startswith(current_color)]
        
        if len(color_matching_actions) != 0:
            action = np.random.choice(color_matching_actions)
        else:
            action = np.random.choice(legal_actions)
        
        return action 
        
    def eval_step(self, state):
        action = self.step(state)
        
        probs = [0 for _ in range(self.num_actions)]
        for i in state['legal_actions']:
            probs[i] = 1/len(state['legal_actions'])

        info = {}
        info['probs'] = {state['raw_legal_actions'][i]: probs[list(state['legal_actions'].keys())[i]] for i in range(len(state['legal_actions']))}
        
        return action, info
```

```python
class UnoAgentNumberMatching(object):
    def __init__(self, num_actions):
        ''' 
        Args:
            num_actions (int): The size of the ouput action space
        '''
        self.use_raw = True
        self.num_actions = num_actions
        
    def step(self, state):
        legal_actions = state['raw_legal_actions']
        state = state['raw_obs']
        current_number = state['target'][-1].split('-')[0]
        
        number_matching_actions = [action for action in legal_actions if current_number in action]
        
        if len(number_matching_actions) != 0:
            action = np.random.choice(number_matching_actions)
        else:
            action = np.random.choice(legal_actions)
        
        return action 
        
    def eval_step(self, state):
        action = self.step(state)
        
        probs = [0 for _ in range(self.num_actions)]
        for i in state['legal_actions']:
            probs[i] = 1/len(state['legal_actions'])

        info = {}
        info['probs'] = {state['raw_legal_actions'][i]: probs[list(state['legal_actions'].keys())[i]] for i in range(len(state['legal_actions']))}
        
        return action, info
```

```python
class UnoAgentSpecialCard(object):
    def __init__(self, num_actions):
        ''' 
        Args:
            num_actions (int): The size of the ouput action space
        '''
        self.use_raw = True
        self.num_actions = num_actions
        
    def step(self, state):
        legal_actions = state['raw_legal_actions']
        state = state['raw_obs']
#         current_type = state['target'][-1].split('-')[0]
        
        special_matching_actions = [action for action in legal_actions if len(action) > 3]
        
        if len(special_matching_actions) != 0:
            action = np.random.choice(special_matching_actions)
        else:
            action = np.random.choice(legal_actions)
        
        return action 
        
    def eval_step(self, state):
        action = self.step(state)
        
        probs = [0 for _ in range(self.num_actions)]
        for i in state['legal_actions']:
            probs[i] = 1/len(state['legal_actions'])

        info = {}
        info['probs'] = {state['raw_legal_actions'][i]: probs[list(state['legal_actions'].keys())[i]] for i in range(len(state['legal_actions']))}
        
        return action, info
```
